```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

## Phase 1
- To complete this phase we merged our movies and ratings 
dataset on movieId. This allowed us to easily generate 
a list of movies watched for each user
- We then used mlextend to perform apriori and association rule mining based off that 2d list which was passed into the transaction encoder

```python
df_ratings = pd.read_csv(r"./data/ml-latest-small/ratings.csv")
df_movies = pd.read_csv(r"./data/ml-latest-small/movies.csv")
```

```python
df_ratings.head()
```

```python
df_movies.head()
```

```python
df_merged = pd.merge(df_ratings, df_movies, on='movieId')
df_merged.head(20)
```

```python
mergedGrouped = df_merged.groupby(['userId'])
dataset = []
#create 2D list of item names per order
for i in range(1,len(mergedGrouped)):
    dataset.append(list(mergedGrouped.get_group(i).title.astype(str)))

te=TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_phase1 = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)
df_phase1
```

```python
df_apriori_p1 = apriori(df_phase1,min_support=0.3,use_colnames=True)
df_apriori_p1
```

```python
rules = association_rules(df_apriori_p1, metric = 'support', min_threshold = 0.01)
rules = rules.sort_values(['confidence','lift'],ascending=False)
rules
```

```python
def findRules(dataframe):
    for i in range(dataframe.shape[0]):
        conf = str(dataframe.iloc[i,5])
        lift = str(dataframe.iloc[i,6])
        ant = str(list(dataframe.iloc[i,0]))
        con = str(list(dataframe.iloc[i,1]))
        print(ant+" --> "+con)
        print("confidence: "+conf+ ", lift: "+lift+"\n")
```

```python
findRules(rules)
```

## Phase 2
- For this phase we created a function that takes a genre as input and creates an encoded dataset that we can use for apriori
- The function works by removing all rows from our merged 
dataset that do not contain the given genre and then 
grouping the remaining data by users so we can extract the movies watched for each user.

```python
'''
Description: Used to create 2d dataset of titles
from a given genre
Input: genre name as defined in movies dataset
Ouput: 2D array of titles based on users
'''

def generateGenreEncoding(genre):
    df_merged_genre = df_merged[df_merged.genres.str.contains(genre)]
    dataset = df_merged_genre.groupby(['userId'])['title'].apply(set)
    te=TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df_phase2 = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)
    return df_phase2
    
```

### comedy

```python
df_comedy = generateGenreEncoding("Comedy")
df_comedy_apriori = apriori(df_comedy, min_support=0.15,use_colnames=True)
rules = association_rules(df_comedy_apriori, metric = 'support', min_threshold = 0.22)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

### Horror

```python
df_horror = generateGenreEncoding("Horror")
df_horror_apriori = apriori(df_horror, min_support=0.15,use_colnames=True)
rules = association_rules(df_horror_apriori, metric = 'support', min_threshold = 0.14)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

### Drama

```python
df_drama = generateGenreEncoding("Drama")
df_drama_apriori = apriori(df_drama, min_support=0.3,use_colnames=True)
rules = association_rules(df_drama_apriori, metric = 'support', min_threshold = 0.01)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

## Phase 3 - Genre Rules
- For phase 3 we realized we needed to turn our genre column into a regular list.
- To accomplish this we joined the genres on the '|' char 
to create one string and then split that string on the same character to create a normal list
- Once we had a list of genres for each row we applied the same process for rule mining.

```python
# grouby userId, join genres on | char, then create lists of genres
df_grules = df_merged.groupby('userId',as_index=False).agg({'genres':'|'.join})
df_grules.genres = df_grules.genres.apply(lambda x: x.split("|")).apply(list)

#encode genre column
dataset = df_grules.genres
te=TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_phase3 = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)
```

```python
#generate rules from encoding
df_genres_apriori = apriori(df_phase3, min_support=.99, use_colnames=True)
rules = association_rules(df_genres_apriori, metric = 'support', min_threshold = 0.996)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

# Phase 4

## Ruleset 1 -- Did people tend to watch good movies, bad movies or both
- To attempt to find the answer we grouped our merged dataset by userId and then took the ratings from each user. This allowed us to generate 
a 2d dataset where each list was a compliation of some users' ratings
- We then applied apriori and the association rule mining mlextend function 

```python
## Ruleset 1 
dataset = df_merged.groupby(['userId'])['rating'].apply(set)
te=TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_r1 = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)

df_r1 = apriori(df_r1,min_support=.75, use_colnames=True)
rules = association_rules(df_r1, metric = 'support', min_threshold = 0.5)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

## Ruleset 2 -- Bad Movie Recommender 
- To find what bad movies people like to watch together we followed the same  proces as our initial movie recommender but removed ratings above 2.5 and lowered our minimum support for generating rules

```python
# make copy of merged dataset
df_low_ratings = df_merged.copy()
```

```python
df_low_ratings = df_low_ratings[df_low_ratings.rating <= 2.5]
dataset = df_low_ratings.groupby(['userId'])['title'].apply(set)
te=TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_r2 = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)

df_r2 = apriori(df_r2,min_support=.025, use_colnames=True)
rules = association_rules(df_r2, metric = 'support', min_threshold = 0.01)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```

## Ruleset 3 -- Horror and Comedy Movie Rules with at Most 3 Star User Ratings

- These movies represent the 'bad' horror and comedy movies that people like to watch the most. The metric for 'bad' is movies that obtained a user rating below 3.0. The process for generating this was similar to above, but with an included metric to filter out movies so that only horror and comedy films are shown. 

```python
df_horror = df_merged.copy()
df_horror = df_horror[df_horror['rating'] < 3.0]
df_horror = df_horror[df_horror.genres.str.contains('Horror')]
df_comedy = df_horror[df_horror.genres.str.contains('Comedy')]
df_horror_comedy = df_horror.merge(df_comedy)
df_horror_comedy = df_horror_comedy.groupby(['userId'])['title'].apply(set)

te=TransactionEncoder()
te_ary = te.fit(df_horror_comedy).transform(df_horror_comedy)

df_horror_comedy = pd.DataFrame(te_ary.astype("int"), columns=te.columns_)

df_horror_comedy = apriori(df_horror_comedy,min_support=.03, use_colnames=True)
rules = association_rules(df_horror_comedy, metric = 'support', min_threshold = 0.03)
rules = rules.sort_values(['confidence','lift'],ascending=False)
findRules(rules)
```
